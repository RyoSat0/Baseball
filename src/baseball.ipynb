Please make sure to **click on the application window first** to focus on it. 

This will ensure that:
- Button clicks work smoothly.
- Typing in the entry fields is responsive.

**Note:** When running the code below, **ensure that Tkinter is supported** on your IDE or code editor;
otherwise, the application window may not be created properly.

```python
import tkinter as tk

class Baseball:
    def __init__(self):
        self.win = tk.Tk()


        width  = int(self.win.winfo_screenwidth() * 0.75)
        height = int(self.win.winfo_screenheight() * 0.75)
        self.win.geometry(f'{width}x{height}+{int(self.win.winfo_screenwidth() * 0.125)}+{int(self.win.winfo_screenheight() * 0.125)}')


        self.numA = ''
        self.numB = ''

        self.frame = tk.Frame(self.win)
        self.frame.place(relx=0.5, rely=0.5, anchor='center')
        self.title = tk.Label(self.frame, text = "BASEBALL", font = ("Times New Roman",100))
        self.title.grid()

        self.askNum = tk.Label(self.frame,text="Player A enter a 3 digit integer with no 0 and no repeated numbers: ")
        self.numEntry = tk.Entry(self.frame)
        self.button = tk.Button(self.frame,text="Next",command=self.goNext)

        self.askNum.grid()
        self.numEntry.grid()
        self.button.grid()

        self.win.mainloop()



    def validate0(self,num):
        for i in range(0,len(num)):
            if num[i] == "0":
                return False

    def validateRepeated(self,num):
        for i in range(0,len(num)-1):
            for j in range(i+1,len(num)):
                if num[i] == num[j]:
                    return False


        for i in range(0,len(num)-1):
            for j in range(i+1,len(num)):
                if num[i] == num[j]:
                    return False

    def validateLength(self,num):
        if len(num) != 3:
            return False

    def fullValidation(self,num):
        if (self.validate0(num) == False or self.validateRepeated(num) == False or self.validateLength(num) == False):
            return False
        else:
            return True



    def goNext(self):
        self.numA = self.numEntry.get()
        self.numEntry.delete(0,tk.END)

        if self.fullValidation(self.numA) == False:
            self.askNum['fg'] = "#FF0000"
            return None

        self.askNum['fg'] = "#000000"
        self.askNum['text'] = "Player B enter a 3 digit integer with no 0 and no repeated numbers: "
        self.button['command'] = self.start
        self.button['text'] = "Start"




    def rightGuess(self,player):
        result = tk.Label(self.win,text=f'Player {player} wins!!',font = ("Times New Roman",100))
        result.grid()
        result.place(x=250,y=250)

        self.guessEntryA['state'] = tk.DISABLED
        self.buttonA['state'] = tk.DISABLED
        self.guessEntryB['state'] = tk.DISABLED
        self.buttonB['state'] = tk.DISABLED



    def checkGuess(self,guess,playerNum):
        ball_count = 0
        strike_count = 0
        for i in range (0,len(guess)):
            for j in range(0, len(playerNum)):
                if guess[i] == playerNum[j] and i != j:
                    ball_count +=1
                if guess[i] == playerNum[j] and i == j:
                    strike_count += 1
        return strike_count, ball_count


    def aTry(self):
        num = self.guessEntryA.get()
        if self.fullValidation(num) == False:
            return None

        strikeNum, ballNum = self.checkGuess(num,self.numB)
        if strikeNum == 0 and ballNum ==0:
            result = "    OUT"
        elif num == self.numB:
            self.rightGuess('A')
            return None
        else:
            result = f"    {strikeNum}S {ballNum}B"

        self.guessHistoryA.append(num + result)
        self.guessA.config(text="\n".join(self.guessHistoryA))

        self.guessEntryA.delete(0, tk.END)

        self.guessEntryA['state'] = tk.DISABLED
        self.buttonA['state'] = tk.DISABLED
        self.guessEntryB['state'] = tk.NORMAL
        self.buttonB['state'] = tk.NORMAL


    def bTry(self):
        num = self.guessEntryB.get()
        if self.fullValidation(num) == False:
            return None

        strikeNum, ballNum = self.checkGuess(num,self.numA)
        if strikeNum == 0 and ballNum ==0:
            result = "    OUT"
        elif num == self.numA:
            self.rightGuess('B')
            return None
        else:
            result = f"    {strikeNum}S {ballNum}B"

        self.guessHistoryB.append(num + result)
        self.guessB.config(text="\n".join(self.guessHistoryB))

        self.guessEntryB.delete(0, tk.END)

        self.guessEntryB['state'] = tk.DISABLED
        self.buttonB['state'] = tk.DISABLED
        self.guessEntryA['state'] = tk.NORMAL
        self.buttonA['state'] = tk.NORMAL

    def start(self):
        self.numB = self.numEntry.get()

        if self.fullValidation(self.numB) == False:
            self.askNum['fg'] = "#FF0000"
            return None

        self.title.destroy()
        self.askNum.destroy()
        self.button.destroy()
        self.numEntry.destroy()



        self.guessHistoryA = []
        self.guessHistoryB = []



        self.playerA = tk.Frame(self.win)
        self.playerB = tk.Frame(self.win)
        self.canvas = tk.Canvas(self.win, height=675, width=5, highlightthickness=0, bd=0)
        self.canvas.create_line(0, 0, 0, 675, width=5, fill="black")


        self.playerA.grid()
        self.playerB.grid()
        self.canvas.grid()
        self.playerA.place(x=140,y=40)
        self.playerB.place(x=680,y=40)
        self.canvas.place(x=540,y=0)

        self.nameLabelA = tk.Label(self.playerA,text="Player A",font = ("Times New Roman",60))
        self.nameLabelB = tk.Label(self.playerB,text="Player B",font = ("Times New Roman",60))
        self.guessA = tk.Label(self.playerA,width=30,height=30,relief="solid",anchor="nw")
        self.guessB = tk.Label(self.playerB,width=30,height=30,relief="solid",anchor="nw")
        self.guessEntryA = tk.Entry(self.playerA,width=10)
        self.guessEntryB = tk.Entry(self.playerB,width=10,state=tk.DISABLED)
        self.buttonA = tk.Button(self.playerA,text="O",command=self.aTry)
        self.buttonB = tk.Button(self.playerB,text="O",state=tk.DISABLED,command=self.bTry)

        self.nameLabelA.grid(row=0, column=0)
        self.nameLabelB.grid(row=0, column=0)
        self.guessA.grid(row=1, column=0)
        self.guessB.grid(row=1, column=0)
        self.guessEntryA.grid(row=2, column=0)
        self.guessEntryB.grid(row=2, column=0)
        self.buttonA.grid(row=2, column=1)
        self.buttonB.grid(row=2, column=1)






test = Baseball()
```
